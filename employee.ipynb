```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('employee_attrition_data - Copy.csv', sep=',')

# Just put the DataFrame as the last line
data.head(10)
```

```python
import pandas as pd

import matplotlib.pyplot as plt



# Now plot the pie chart
data["Gender"].value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.ylabel("")
plt.title("Gender Count of Employees")
plt.show()
```

```python
x=data[["Years_at_Company","Satisfaction_Level","Average_Monthly_Hours"]]
y=data["Salary"]
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
```

```python
x_train
```

```python
import joblib
joblib.dump(scalar,"scalar.pkl")
```

```python
x_test=scalar.transform(x_test)
```

```python
x_test
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error

def results(prediction):
  print("Mean absolute error on model is {}",format(mean_absolute_error(y_test,prediction)))
  print("Root mean squared error on model is {}",format(np.sqrt(mean_squared_error(y_test,prediction))))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictionlr=lr.predict(x_test)
```

```python
results(predictionlr)
```

```python
from sklearn .model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
svrmodel=SVR()
```

```python
param_gridsvr = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf'],
    'degree': [2, 3, 4]
}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(x_train,y_train)
```

```python
gridsvr.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rfmodel = RandomForestRegressor()
```

```python
param_gridrfr = {
    "n_estimators": [2, 3, 4, 5, 6],  # Use a list instead of a set
    "max_depth": [5, 10, 15]
}
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='viridis')
plt.show()
```

```python
gridrfr = GridSearchCV(estimator=rfmodel, param_grid=param_gridrfr, cv=5)
```

```python
gridrfr.fit(x_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
predictionofrfr = gridrfr.predict(x_test)

```

```python
results(predictionofrfr)
```
